```python

import heapq

def decodeMessage(message: str) -> str:
    output = ""
    heap = []
    for char in message:

        if char.isdigit():
          heap.pop()
        else:
          heap.append( char)


    if len(heap) == 0:
      return ''
    else:
      return ''.join(heap)
```

```python
message = "abc"
print(decodeMessage(message))
```

```python
def function3(n):
    i = 1
    counter = 0
    while i * i < n:
        i += 1
        counter += 1
    return counter

def function5(n):
    i = 2
    counter = 0
    while i < n:
        i = i * i
        counter += 1
    return counter

print(function3(1000))
print(function5(1000))
```
