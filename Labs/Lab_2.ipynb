```python
def isAnagram(word1, word2):

    if len(word1) == len(word2):
        seen = {}
        seen2 = {}
        for letter in range(len(word1)):

          let_1 = word1[letter]
          let_2 = word2[letter]

          if let_1 in seen:
              seen[let_1] += 1
          else:
              seen[let_1] = 1

          if let_2 in seen2:
              seen2[let_2] += 1
          else:
              seen2[let_2] = 1


        for unique_letter, count in seen.items():

            if unique_letter in seen2 and seen2[unique_letter] == count:
                continue
            else:
                return "No, they are not Anagrams"

        return "Yes, they are Anagrams"
    else:
        return "No, they are not Anagrams"





```

```python
isAnagram('rat', 'car')
```

```python
isAnagram('anagram', 'nagaram')
```
