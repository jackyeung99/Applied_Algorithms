```python
import random 
import matplotlib.pyplot as plt
```

```python

def compute_range(n):
    count = 0
    for x in range(n):
        y = random.uniform(0,1)

        if (x*2) + (y*2) < 1:
            count += 1

    return 4*count/n


n_range = range(100, 10000, 100)
vals = [compute_range(n) for n in n_range]

```

```python
plt.plot(n_range, vals)
```
