```python
import random 
import matplotlib.pyplot as plt
```

```python

def compute_range(n):
    count = 0
    for _ in range(n):
        y = random.uniform(-1.0,1.0)
        x = random.uniform(-1.0,1.0)
        if (x**2) + (y**2) < 1:
            count += 1 

    return (4*count)/n

```

```python
n_range = range(100, 100000, 100)
vals = [compute_range(n) for n in n_range]
plt.plot(n_range, vals)
```

Observations:

This simulations is an example of a monte carlo algorithm becasue of the inability to guarantee a correct solution. For this case we take a random x and y coordinate value between -1 and 1 and if x^2 + y^2 > 1 than that is incorrect output. The simulation is ran n times and the percentage of failures is marked. This is reflected within the graph were we notice the failure rate on the y value and the amount of repetitions on X. The results of this simulation is a stabilization of error rate as we increase repetitions, this is shown in the decreasing variance as X increases.
